#### while

```python
x = 0
while 2 <= 10:
    print(x, 'is less than 10')
    x += 2
#     x = x+2
```

```python
x = list(range(11))
x
```

```python
a = 0
status = True
while status:
    a += 1
    print(a)
    if a >= 10:
        status = False
```

```python
#break
n = 1
while n < 5:
    if n == 3:
        break
    n += 1
    print(n)

print('output',n)
```

```python
list(range(4))
```

```python
#continue
n = []
for i in range(4):
    if i%2 == 0:
        continue
    n.append(i)
print(n)
```

```python
n = 1
while n<5:
    print(n)
    n += 1
else:
    print('end')
```

```python
def add(n1, n2):
    print(n1, n2)
    return n1 + n2
```

```python
def exam(a, b):
    n = 0
    m = []
    for i in a:
        n += i*b
        m.append(n)
    return m
```

```python
exam([1, 2, 3, 5], 5)
```
